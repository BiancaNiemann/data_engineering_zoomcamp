#### Question 3. 

How many rows are there for the Yellow Taxi data for all CSV files in the year 2020?

```python
query = """
SELECT count(*) 
FROM `kestra-sandbox-485712.zoomcamp.yellow_tripdata` 
WHERE DATE(tpep_pickup_datetime) >= '2020-01-01'
  AND DATE(tpep_pickup_datetime) <= '2020-12-31'
"""

client.query(query).to_dataframe()
```

#### Question 4. 
How many rows are there for the Green Taxi data for all CSV files in the year 2020?

```python
query = """
SELECT count(*) 
FROM `kestra-sandbox-485712.zoomcamp.green_tripdata` 
WHERE TIMESTAMP_TRUNC(lpep_pickup_datetime, YEAR) = TIMESTAMP("2020")
"""

client.query(query).to_dataframe()
```
